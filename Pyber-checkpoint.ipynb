```python
#Dependencies

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.cm as cm
```

```python
#Datafiles - Read data files and check raw data

ride = pd.read_csv("C:/Users/kapali_s/Documents/SMU/Homeworks/Assignment_5/ride_data.csv")
city = pd.read_csv("C:/Users/kapali_s/Documents/SMU/Homeworks/Assignment_5/city_data.csv")
```

```python
#Read data for city
city
```

```python
#Combine the data for city and ride
#Outer join so all information is pulled

city_ride= pd.merge(city, ride, how="outer", on="city")
city_ride.head()
```

```python
#Number of rides per city - count the # of time the city occurs and create a dataframe
#Count how many times each city occurs
city_num_rides = city_ride["city"].value_counts()
city_num_rides
```

```python
#Create a dataframe for the number of rides
num_rides = pd.DataFrame(city_num_rides.reset_index().values,columns=["city","No.of Rides"])
num_rides
```

```python
#Calculate average fare per city - Groupby city and average the fare
#Group by city
city_group = city_ride.groupby(["city"])
city_group.head()

#Calculate the mean and reset index to create a data frame
city_avg_fare = city_group["fare"].mean().reset_index()
city_avg_fare

#Rename column fare to Avegare Fare
city_avg_fare = city_avg_fare.rename(columns={"fare":"Average Fare"})
city_avg_fare


```

```python
#Line up city and type - Assign type for each city
#Groupby city ; Pick only columns city and type and drop duplicates

city_group = city_ride.groupby(["city"])
city_group_type = city_group[["city",'type']].head().drop_duplicates()
city_group_type 
```

```python
#Calculate number of drivers per city - Groupby city and count drivers
#Group by city
city_group = city_ride.groupby(["city"])
city_group.head()

#Calculate the no. of drivers and reset index to create a data frame
city_drivers_count = city_group["driver_count"].count().reset_index()
city_drivers_count
```

```python
#Merge the dataframes for no. of rides and average fare
ride_share_df = pd.merge(num_rides,city_avg_fare, on="city", how="outer")
ride_share_df

#Merge the dataframes to include type
ride_share_df1 = pd.merge(ride_share_df,city_group_type, on="city", how="outer")
ride_share_df1

#Merge the dataframes to include driver count
ride_share_df2 = pd.merge(ride_share_df1,city_drivers_count, on="city", how="outer")
ride_share_df2

ride_share_df2
```

# Total fares by city type

```python
#Calculate percent by type

#Take only columns type and fare
city_fare = city_ride[["fare","type"]]
city_fare

#Add percent fare
#Calculate total fare
total_fare = city_fare["fare"].sum()
total_fare

#Group by type
city_fare_type = city_fare.groupby(["type"])
#Total fare for each city and reset index
city_fare_type_total = city_fare_type["fare"].sum().reset_index()
city_fare_type_total

#Calculate percentage for each type
city_fare_type_total["Percentage Fare"] = (city_fare_type_total["fare"] / total_fare)*100
city_fare_type_total

# Build Pie Chart
labels = city_fare_type_total["type"]
sizes = city_fare_type_total["Percentage Fare"]
colors = ["red", "orange", "lightskyblue"]
explode = (0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)


```
