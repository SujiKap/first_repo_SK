```python
#Dependencies

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Datafiles - Read data files and check raw data

ride = pd.read_csv("C:/Users/kapali_s/Documents/SMU/Homeworks/Assignment_5/ride_data.csv")
city = pd.read_csv("C:/Users/kapali_s/Documents/SMU/Homeworks/Assignment_5/city_data.csv")
```

```python
#Read data for ride
ride.head()
```

```python
#Read data for city
city.head()
```

```python
#Combine the data for city and ride
#Join so all information is pulled

city_ride = pd.merge(ride, city, how="left", on=["city", "city"])
city_ride.head()
#city_ride[city_ride["city"] == "Williamsstad"]
```

```python
#Number of rides per city - count the # of time the city occurs and create a dataframe
#Count how many times each city occurs
city_num_rides = city_ride["city"].value_counts()
city_num_rides
```

```python
#Create a dataframe for the number of rides
num_rides = pd.DataFrame(city_num_rides.reset_index().values,columns=["city","No.of Rides"])
num_rides
```

```python
#Calculate average fare per city - Groupby city and average the fare
#Group by city
city_group = city_ride.groupby(["city"])
#city_group.head()

#Calculate the mean and reset index to create a data frame
city_avg_fare = city_group["fare"].mean().reset_index()

#Rename column fare to Avegare Fare
city_avg_fare = city_avg_fare.rename(columns={"fare":"Average Fare"})
city_avg_fare
```

```python
#Line up city and type - Assign type for each city
#Groupby city ; Pick only columns city and type and drop duplicates

city_group = city_ride.groupby(["city"])
city_group_type = city_group[["city",'type']].head().drop_duplicates()
city_group_type 
```

```python
#Calculate number of drivers per city - Groupby city and count drivers
#Group by city
city_group = city_ride.groupby(["city"])
#city_group.head()

#Calculate the no. of drivers and reset index to create a data frame
city_drivers_count = city_group["driver_count"].count().reset_index()
city_drivers_count
```

```python
#Merge the dataframes for no. of rides and average fare
ride_share_df = pd.merge(num_rides,city_avg_fare, on="city", how="outer")
ride_share_df

#Merge the dataframes to include type
ride_share_df1 = pd.merge(ride_share_df,city_group_type, on="city", how="outer")
ride_share_df1

#Merge the dataframes to include driver count
ride_share_df2 = pd.merge(ride_share_df1,city_drivers_count, on="city", how="outer")
ride_share_df2

```

# Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

#Create separate dataframes for each city type
city_type_urban = city_ride.loc[(city_ride["type"] == "Urban")]
city_type_rural = city_ride.loc[(city_ride["type"] == "Rural")]
city_type_suburban = city_ride.loc[(city_ride["type"] == "Suburban")]

#Caulate rides,fares and driver counts for urban city type
urban_city_group = city_type_urban.groupby(["city"])

u_x = urban_city_group.count()["ride_id"]
u_y = urban_city_group["fare"].mean()
u_s = urban_city_group["driver_count"].mean()

#Caulate rides,fares and driver counts for rural city type
rural_city_group = city_type_rural.groupby(["city"])

r_x = rural_city_group.count()["ride_id"]
r_y = rural_city_group["fare"].mean()
r_s = rural_city_group["driver_count"].mean()

#Caulate rides,fares and driver counts for suburban city type
suburan_city_group = city_type_suburban.groupby(["city"])

s_x = suburan_city_group.count()["ride_id"]
s_y = suburan_city_group["fare"].mean()
s_s = suburan_city_group["driver_count"].mean()

# Build the scatter plots for each city types
fig, ax = plt.subplots()

urban_plot= ax.scatter(u_x, u_y, s=u_s*10, color=["lightcoral"], edgecolors = 'black', alpha = 0.75, label="Urban" ,linewidth = 1.5)
rural_plot= ax.scatter(r_x, r_y, s=r_s*10, color=["gold"], edgecolors = 'black', alpha = 0.75, label="Rural",linewidth = 1.5)
suburban_plot= ax.scatter(s_x, s_y, s=s_s*10, color=["lightskyblue"], edgecolors = 'black', alpha = 0.75, label="Suburban",linewidth = 1.5)

#Name the labels and title
ax.set_xlabel("Total Number of Rider (Per City)")
plt.ylabel("Average Fare($)")
plt.title("Pyber Ride Sharing Data (2016)")

# Incorporate the other graph properties
ax.grid()

# Create a legend
legend = plt.legend(loc="upper right", fontsize = 10, title= "City Types")

legend.legendHandles[0]._sizes = [35]
legend.legendHandles[1]._sizes = [35]
legend.legendHandles[2]._sizes = [35]

# Incorporate a text label regarding circle size
plt.text(44,35,"Note: \n Circle size correlates with driver count per city.", fontsize = 12)

# Save Figure

plt.savefig("C:/Users/kapali_s/Documents/SMU/Homeworks/Assignment_5/Pyber Ride Sharing.png",bbox_inches=("tight"))
plt.show()
plt.tight_layout
```

# Total fares by city type

```python
#Calculate percent by type

#Take only columns type and fare
city_fare = city_ride[["fare","type"]]
city_fare

#Add percent fare
#Calculate total fare
total_fare = city_fare["fare"].sum()
total_fare

#Group by type
city_fare_type = city_fare.groupby(["type"])
#Total fare for each city and reset index
city_fare_type_total = city_fare_type["fare"].sum().reset_index()
city_fare_type_total

#Calculate percentage for each type
city_fare_type_total["Percentage Fare"] = (city_fare_type_total["fare"] / total_fare)*100
city_fare_type_total

#Build Pie Chart
labels = city_fare_type_total["type"]
sizes = city_fare_type_total["Percentage Fare"]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0, 0, 0.1)

city_fare_pie = plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=140)


ax.axis("equal")

#Label the chart
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("C:/Users/kapali_s/Documents/SMU/Homeworks/Assignment_5/Total Fares By City.png")
plt.show()

```
